### Q1:
`time_to_delivery` - это метрика, которая считает сколько проходит между созданием (`created_at`) и получением заказа (`delivered_at`)

Посчитайте для каждого города средний `time_to_delivery` и отсорируйте их - от худшего к лучшему

```python
from IPython.core.display import display, HTML
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "all"
display(HTML("<style>.container { width:85% !important; }</style>"))
```

```python
city_dates = pd.read_sql_query("""SELECT city_name,
                                         ROUND(AVG((JULIANDAY(delivered_at) - JULIANDAY(created_at)) * 24)) AS time_to_delivery
                                    FROM cities
                               LEFT JOIN orders USING(city_id)
                                   GROUP BY city_name
                                   ORDER BY time_to_delivery""", con)
city_dates
```

#### Овтет дан в часах


### Q2: 
А сколько в среднем проходит между тем как пользователь получил первый заказ, и сделал следующий заказ?

```python
mean_timedelta_delivery = pd.read_sql_query("""SELECT AVG(ROUND((SELECT JULIANDAY(MIN(delivered_at)) * 24
                                                                   FROM orders
                                                                  GROUP BY user_id) - (SELECT JULIANDAY(created_at)
                                                                                         FROM orders
                                                                                        GROUP BY user_id
                                                                                        ORDER BY created_at
                                                                                        LIMIT 1 OFFSET 1) * 24)) AS mean
                                                 FROM orders""", con)

mean_timedelta_delivery
```

```python
pd.read_sql_query("""
    WITH q1 AS(
        SELECT MIN(JULIANDAY(delivered_at) * 24) -
               LEAD(JULIANDAY(created_at) * 24)
               OVER(ORDER BY JULIANDAY(created_at) * 24) AS time
          FROM orders
         GROUP BY user_id
    )
    SELECT ROUND(AVG(time)) AS mean FROM q1""", con)
```

#### Ответ: В среднем проходит 105 часов

### Q3: 

Посчитайте сколько всего пользователей в Москве имеют средний чек выше 2000 рублей (в % от всех пользователей, у которых был хотя бы один заказ в Москве)

##### Т.к. в базе нет колонки в каком городе проживает человек, а есть только колонка, в которой указано откуда(или куда) был доставлен товар, буду считать, что "пользователи в Москве", это те пользователи, которые хоть раз что-то заказывали в Москве

Если в задании подразумевалось то, что нужно определить живет ли человек в Москве, это будет, как мне кажется, немного некорректно, т.к. Если мы будем считать что человек живет в Москве, если у него все покупки из/в Москву, он мог быть просто проездом в Москве и заказать только 1-ин товар и после этого перестал пользоваться сервисом/маназином.

Если мы будем считать Москвичами тех, людей, которые большую часть своих покупок совершали в Москве, это тоже не совсем корректно, тк у нас много покупателей всего с одним заказом и тогда мы снова возращаемся к первому пункту.

```python
pd.read_sql_query("""WITH 
                       q1 AS (SELECT COUNT(DISTINCT user_id) AS moscow_more_2000
                                FROM orders
                               WHERE user_id IN(SELECT user_id
                                                  FROM cities LEFT JOIN orders USING(city_id)
                                                 WHERE city_name = 'Moscow'
                                                 GROUP BY user_id
                                                HAVING AVG(value_rub) > 2000
                                               )
                             ),
                       q2 AS (SELECT COUNT(DISTINCT user_id) AS moscow
                                FROM cities LEFT JOIN orders USING(city_id)
                               WHERE city_name = 'Moscow'
                             )
                   SELECT ROUND(CAST(moscow_more_2000 AS FLOAT) / CAST(moscow AS FLOAT) * 100) AS procent FROM q1, q2""", con)
```

#### Ответ: 39%


### Q4:
Посчитайте сколько часов в среднем проходит между заказами одного и того же пользователя?

```python
df_orders.groupby('user_id') \
         .agg({'created_at': 'count'}) \
         .value_counts()

df_orders.user_id.nunique()
```

#### Ответ с учетом тех пользователей, которые совершили одну и более покупки:

```python
pd.read_sql_query("""SELECT ROUND((MAX(JULIANDAY(created_at)) - MIN(JULIANDAY(created_at))) / (COUNT(created_at) - 1)  * 24) AS hours
                       FROM orders
                      GROUP BY user_id""", con)
```

* У тех, кто совершил только одну покупку, среднее = NaN(0)

#### Ответ с пользователями которые совершили более одной покупки:

```python
pd.read_sql_query("""SELECT ROUND((MAX(JULIANDAY(created_at)) - MIN(JULIANDAY(created_at))) / (COUNT(created_at) - 1)  * 24) AS hours
                       FROM orders
                      GROUP BY user_id
                     HAVING COUNT(created_at) > 1""", con)
```

#### Среднее кол-во часов между заказами среди всех пользователей у которых кол-во заказов больше одного:

```python
pd.read_sql_query("""WITH q AS (SELECT ROUND((MAX(JULIANDAY(created_at)) - MIN(JULIANDAY(created_at))) / (COUNT(created_at) - 1)  * 24) AS hours
                                  FROM orders
                                 GROUP BY user_id
                                HAVING COUNT(created_at) > 1
                               ) SELECT ROUND(AVG(hours)) AS mean FROM q""", con)
```
