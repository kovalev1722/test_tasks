# Задача 1

# 1.1. Подготовка данных

```python
clients = pd.read_parquet('clients.parquet')
shows = pd.read_parquet('shows.parquet')
sales = pd.read_parquet('sales.parquet')
products = pd.read_parquet('products.parquet')
```

```python
shows.calday.dt.isocalendar().week.value_counts()
```

#### Пополнение таблиц с показами и продажами, информацией о клиентах и названиями продуктов

```python
sales = (clients.merge(sales, how='right')) \
    .merge(products, how='left', left_on = 'product_id', right_on='id')

shows = (clients.merge(shows, how='right')) \
    .merge(products, how='left', left_on = 'product_id', right_on='id')
```

#### Разделение данных по датам.

* `sales_before_ab` и `sales_before_ab` - до AB-теста;
* `sales_ab`, `shows_ab` - во время AB-теста.

```python
sales_before_ab = sales.loc[(sales.calday < '2021-02-24') & ((sales.product_name == 'deposit_4') | (sales.product_name == 'insurance'))]
shows_before_ab = shows.loc[(shows.calday < '2021-02-24') & ((shows.product_name == 'deposit_4') | (shows.product_name == 'insurance'))]

sales_ab = sales.loc[(sales.calday >= '2021-02-24') & (sales.calday <= '2021-03-12') & \
                    ((sales.product_name == 'deposit_4') | (sales.product_name == 'insurance'))]
shows_ab = shows.loc[(shows.calday >= '2021-02-24') & (shows.calday <= '2021-03-12') & \
                    ((shows.product_name == 'deposit_4') | (shows.product_name == 'insurance'))]
```

#### Таблица данных AB-теста

Содержимое таблицы **`df_ab`:**

* `user_count` - кол-во уникальных пользователей, которые купили товар/увидели товар;
* `sales` и `shows` - общее кол-во покупок/показов

$$convs = \frac{sales}{shows} * 100%$$

```python
df_ab = pd.DataFrame(sales_ab.groupby('client_id') \
    .agg({'group_type': 'max'}) \
    .rename(columns={'group_type': 'users_count_sales'}) \
    .users_count_sales.value_counts())

df_ab['users_count_shows'] = shows_ab.groupby('client_id') \
    .agg({'group_type': 'max'}) \
    .group_type.value_counts()

df_ab['sales'] = sales_ab.group_type.value_counts()
df_ab['shows'] = shows_ab.group_type.value_counts()

df_ab['convs'] = round(df_ab.sales / df_ab.shows * 100, 2)

df_ab
```

# 1.3. Визуализация

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
sales_per_week = sales.calday.dt.isocalendar().week.value_counts()
shows_per_week = shows.calday.dt.isocalendar().week.value_counts()
convs_per_week = np.divide(sales_per_week, shows_per_week)

per_week = pd.DataFrame({
    'sales': sales_per_week,
    'shows': shows_per_week,
    'convertion = 17%': convs_per_week * 100
})

sale_client = sales.groupby('client_id').client_id.count().sort_values(ascending=False).value_counts()
```

```python
fig, axs = plt.subplots(2, 2)
sns.set(rc={'figure.figsize':(30, 20)})

fontdict = {'fontsize': 10,
            'fontweight' : 10}

sns.barplot(data=df_ab, x=df_ab.index, y=df_ab.convs, palette='crest', ax=axs[0, 0])
axs[0, 0].set_title('Conversion during AB-test', fontsize=20)
axs[0, 0].set_xlabel("Group type", fontsize=20)
axs[0, 0].set_ylabel("Conversion %", fontsize=20)
  
sns.lineplot(data=per_week, ax=axs[0, 1])
axs[0, 1].set_title('Sales and shows', fontsize=20)
axs[0, 1].set_xlabel("Week", fontsize=20)
axs[0, 1].set_ylabel("Quantity", fontsize=20)
  
sns.countplot(data=sales, y=sales.product_name, order=sales.product_name.value_counts().index, palette='crest', ax=axs[1, 0])
axs[1, 0].set_title('TOP products', fontsize=20)
axs[1, 0].set_xlabel("Quantity of sales", fontsize=20)
axs[1, 0].set_ylabel("Product", fontsize=20)
  
sns.barplot(y=sale_client.index, x=sale_client.values, hue_order=sale_client.index, palette='crest', orient='h', ax=axs[1, 1])
axs[1, 1].set_title('Quantity of sales per clients', fontsize=20)
axs[1, 1].set_xlabel("Quantity of clients", fontsize=20)
axs[1, 1].set_ylabel("Quantity of sales", fontsize=20)
```

# 1.2. Анализ AB-теста.

### Мощность теста

Чтобы увидеть вероятность статистически значимых различий там, где они действительно существуют, вычисляется мощность теста.

Меня будет устраивать мощность выше 80% (в 80 случаев из 100 я смогу увидеть разницу там, где она есть)

```python
import statsmodels.stats.power as smp
```

```python
chipower = smp.GofChisquarePower()
```

```python
def chi2_effect_size(convs):
    return np.sqrt(((convs[1]-convs[0])**2) / convs[1])

convs = df_ab.sales / df_ab.shows

power = round(chipower.solve_power(
    effect_size=chi2_effect_size(convs),
    nobs=df_ab.shows[1],
    alpha=0.05,
    power=None) * 100, 2)

print(f'Мощность = {power}%')
```

Мощность 99.5%, на порядок больше, чем меня бы устраивало. Это означает, что мы можем не переживать насчет дальнейшей проверки на стат. значимость различий метрик (мы не упустим стат. значимые различия если они сущетвуют).

### Критерий  Хи-квадрат

Чтобы определить статистическую значимость различий конверсий наших групп, применяю критерей **$\chi^2$**.

```python
import statsmodels.stats.proportion as proportion
```

```python
chi2stat, p_val, table = proportion.proportions_chisquare(df_ab.sales, df_ab.shows)
```

Уровень значимости `alpha` задаю стандартный, равный **0.05**.

```python
alpha = 0.05

print(f'alpha = {alpha}')
print(f'p_val = {round(p_val, 4)}')
print(f'\np_val < alpha is {p_val < alpha}')
```

`p_value` = 0.0007, это означает, что только в **0.07%** случаев мы можем случайно получить такие или ещё более выраженные различия между тестовой и контрольной группой.
