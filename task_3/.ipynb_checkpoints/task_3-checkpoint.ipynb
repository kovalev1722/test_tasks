## Какие у неё пользователи? Какие цели и задачи?

### Какие у неё пользователи:

1) **Аудитория в возрасте 16-60**. 
* 16 - возраст, когда молодые люди уже довольно часто начинают пользоваться метрополитеном, чтобы куда-то добраться.
* 60 - примерная оценка, что люди более старшего возраста, в нынешней эпидимиологической ситуации, стараются избежать нахождения в людных местах(в особенности метро)

2) **Распределение полов**:

* 56% - женщины
* 44% - мужчины

3) **Доходы пассажиров**: 

* 67% пассажиров имеют доход средний или выше среднего
* 31% - пассажиров имеют доход ниже среднего
* 2% не определены

4) Более вероятно, что пассажиры живут относительно далеко от центра города (большинство пассажиров живут не в ЦАО).

### Какие цели и задачи:

1) Ст. метро находится в одном месте с одноименным вокзалом, в связи с этим, можно сделать вывод, что большое кол-во людей заходят на эту станцию с электричек, которые приехали с западного направления. Из-за этого довольно сильно увеличивается пассажиропоток. Так же на это влияет большой торговый центр.

2) Выезд из/в город(а) с после или перед пересадкой на/с электричек.

3) Большой пассажиропоток, обусловлен тем, что Киевскую можно назвать "перевалочным пунктом", в котором пересекаются сразу 3 ветки метро и вокзал, которые ведут как в центр города, кольцо и западное направление.

4) Большая часть людей использует его(метро) для поездки на работу, в частности ст. метро "Киевская".

## Какие метрики ты бы использовал, чтобы описать ее с точки зрения данных? Для чего они нужны? Как связаны между собой? Какие выводы и действия ты можешь сделать смотря на эти метрики?

1) **DAU, MAU, HAU**(hoursly active users) - чтобы понимать и видить динамику того, какое кол-во людей пользуются/проезжают эту станцию метро.
* Его рост даст не особо много информации, тк для этих метрик, колебания в связи с выходными/праздниками/сезоном отпусков нормальное явление.
* Но сильное снижение пассажиропотока может дать информацию Мос. гор. трансу, что маршруты(например наземного транспорта) не справляются с задачей доставки людей до этой станции метро или прилежащих к ней веток(синяя/желтая).
* HAU - Чтобы была возможность мониторить ситуацию на этой станции в данный момент(если произошло что-то критичное в метро).

2) **CR** - полезным действием будем считать покупку билета или пробитие билета у турникетов на этой станции.
* Эта метрика может помочь понять какое кол-во пассажирова была проездом на этой станции, а какая села в метру именно на ней.

3) **ARPU, ARPPU, Средний чек** - наблюдение за прибылью с каждого пользователя и по каждому заказу.
* Помогает наблюдать и влиять на кол-во потраченых денег пассажиров, по средствам персональных скидок, предложений(т.к. сейчас начинают внидрять персонализированные карты "Тройка", это будет намного проще сделать).

4) **Churn Rate** - если появится станция которая заберет на себе часть пассажиропотока, мы увидим отток пассажиров.

5) **LTV** - если мы можем предсказать какое примерно кол-во денег пользователь потратит, мы можем спровоцировать его на покупку лишних поездок, которые он будет брать на будущее и в итоге их не использует.

6) **Clicks, View, CTR** - клики, просмотры, их соотношение. Для рекламодателей, которые берут рекламу у метро, поможет понять как изменить дизайн сайта, чтобы повысить CTR.

7) **ROI**(Return On Investment) - мониторинг окупаемости акций и специальных предложений для пассажиров. Реклама не рассматривается, тк метро в ней особо не нуждается(хотя можно проводить пропагандистскую кампанию отказа от своего транспорта и пересадки на общественный(в частности метро))

8) **Время на станции, среднее время между покупкой билета и проходом через турникеты.**
* Время на станции - помогает анализировать какое количество людей будет находиться в определенный промежуток времени на станции, чтобы включить/выключить эскалаторы, оптимизировать передвижение по станции.
* Время между покупкой и проходом - с его помощью можно понять нсколько сильная загруженность на входе в среднем и какое кол-во людей покупает поездки заранее, а какое каждый раз, когда заходит в метро. Помогает выбрать оптимаьное кол-во касс, автоматов.

**Это те метрики, которые приходят сразу на ум.**

# Задание 2

В компании QIWI есть несколько продуктов, но в каждом из них нужна отдельная идентификация.
Продуктовый менеджер думает о разработке возможности единой идентификации во всех продуктах
и просит нас посмотреть, какое кол-во пользователей за последнюю неделю проходили идентификацию
в других продуктах (не в QIWI-кошельке), но при этом уже встречались с такими же паспортными
данными (серия и номер) и фио в кошельке?

(*) идентификация: процесс, в котором пользователь указывает и подтверждает свои паспортные данные и другую персональную информацию

Таблицы:

identification

- user_id - строка, не null
- product_id - целое, не null
- identification_date - time, не null

(user_id - первичный ключ; QIWI-кошелек - это product_id = 1)

table_with_personal_data
- user_id - строка, не null
- pers_key_id - целое, не null
- pers value - строка, не null

user_id, pers_key_id - составной первичный ключ

pers_key_id равный 1, 2, 3, 4, 5 означают, соответственно, фамилию, имя, отчество, серию паспорта
и номер паспорта. Список значений pers_key_id насчитывает сотни наименований.

Напишите SQL-запрос

### Подсказка ко второй задаче 

Если пользователь зарегистрировался в разных продуктах, то user_id будет разный.


### 1 вариант решения

```python
"""
WITH q_1(user_id, full_info) AS (
SELECT user_id, GROUP_CONCAT(pers_value SEPARATOR ' ') as full_info
  FROM identification JOIN table_with_personal_data USING(user_id)
  WHERE identification_date BETWEEN NOW() - INTERVAL 1 week AND NOW()
 GROUP BY user_id
)
SELECT COUNT(DISTINCT full_info)
  FROM q_1
 GROUP BY full_info
HAVING COUNT(full_info) > 1;
"""
```
